```python
data_path='../../data/home-credit/'

app_train = pd.read_csv(f'{data_path}application_train.csv')
app_test = pd.read_csv(f'{data_path}application_test.csv')
bureau = pd.read_csv(f'{data_path}bureau.csv')
bureau_balance = pd.read_csv(f'{data_path}bureau_balance.csv')
```

## Processamento da tabela 'application'

### Tratamento de Anomalias em DAYS_EMPLOYED

```python
app_train['DAYS_EMPLOYED_ANOM'] = app_train['DAYS_EMPLOYED'] == 365243
app_test['DAYS_EMPLOYED_ANOM'] = app_test['DAYS_EMPLOYED'] == 365243
app_train['DAYS_EMPLOYED'].replace(365243, np.nan, inplace=True)
app_test['DAYS_EMPLOYED'].replace(365243, np.nan, inplace=True)
```

```python
days_cols = [col for col in app_train.columns if 'DAYS' in col]
for col in days_cols:
    app_train[col] = abs(app_train[col])
    app_test[col] = abs(app_test[col])
```
